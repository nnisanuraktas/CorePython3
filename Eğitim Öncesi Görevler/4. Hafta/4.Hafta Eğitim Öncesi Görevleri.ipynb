Lütfen numarasını öğrenmek istediğiniz kişinin adını girin: Ahmet

# Problem 3

Kullanıcıya hangi öğrencinin notlarını görmek istediğini sorun. Öğrencinin notu görüntülendiğinde program sonunda şöyle bir çıktı elde etmelisiniz:

```
Lütfen notlarını görmek istediğiniz öğrencinin adını girin: Mehmet

Mehmet isimli öğrencinin      1.Sınav Notu:72
                              2.Sınav Notu:66
                              3.Sınav Notu:48

````python
notlar= {
    "Nisa" : [90,27,80],
    "Açelya" : [90,100,50],
    "Mert" : [60,40,75]
}
ogrenci=input("isim yaz")
ortalama= (notlar[ogrenci][0] + notlar[ogrenci][1] + notlar[ogrenci][2]) / 3
print(f"""{ogrenci} isimli öğrencinin 1. Sınav Notu : {str(notlar[ogrenci][0])}
                                    2. Sınav Notu:  {str(notlar[ogrenci][1])}
                                    3. Sınav Notu:  {str(notlar[ogrenci][2])}
        Not ortalaması :{str(ortalama)}""")
        
````
